```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
T=np.arange(100,850,50)
T
Cp=[40.144,41.826,43.396,44.860,46.222,47.489,48.665,49.755,50.764,51.698,52.559,53.353,54.084,54.755,55.371]
```

```python
T=T+273.15
y=Cp
t=T/1000
n=len(T)
X=np.array([np.ones(n),t,t**2,t**3,1/t**2]).T
X
beta=X.T.dot(X)
beta=np.linalg.inv(X.T.dot(X)).dot(X.T.dot(y))
beta
Cp_calc=X.dot(beta)
Cp_calc
```

```python
plt.scatter(T,y,color="navy")
plt.plot(T,Cp_calc)
```

```python
dT=0.1
T=np.arange(100,800,dT)
t=T/1000
n=len(T)
X=np.array([np.ones(n),t,t**2,t**3,1/t**2]).T
Cp_calc=X.dot(beta)
plt.plot(T,Cp_calc)
plt.show()
```

```python
H=np.cumsum(Cp_calc*dT)
plt.plot(T,H)
plt.xlabel("T(K)")
plt.ylabel("delta_H(J/mol)")
plt.show()
```

```python
S=np.cumsum(Cp_calc/T*dT)
plt.plot(T,S)
plt.xlabel("T(K)")
plt.ylabel("delta_H(J/mol)")
plt.show()
```

```python
G=H-T*S
plt.plot(T,G)
plt.xlabel("T(K)")
plt.ylabel("delta_G(J/mol)")
plt.show()
```
